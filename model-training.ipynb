```python
# cat codes 
fin=[]
for i in columns:
    c = df[i].astype('category')
    d = dict(enumerate(c.cat.categories))
    df[i] = df[i].astype('category').cat.codes
    temp=df[i].map(d)
    fin.append(i)
    fin.append(d)
#print(temp)
```

```python
x=df.drop('age',axis=1)
y=df['age']
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(x,y ,random_state=104,test_size=0.2,shuffle=True)
```

# Given Model

```python
import tensorflow as tf
model=tf.keras.Sequential()
bias=tf.keras.initializers.HeNormal()
model.add(tf.keras.layers.Dense(1,kernel_initializer="he_normal",use_bias=True,bias_initializer=bias,input_shape=(64,)))
model.summary()
model.compile(optimizer=tf.keras.optimizers.Adam(0.1),loss=tf.keras.losses.MeanSquaredError(),metrics=[tf.keras.metrics.RootMeanSquaredError()])
learning=model.fit(X_train,y_train,epochs=100,batch_size=16,verbose=True)
```

```python
import matplotlib.pyplot as plt
def plot(learning):
  plt.plot(learning.history['loss'])
  plt.xlabel("epochs")
  plt.ylabel("Cost")

  plt.show()
  plot(learning)
```

```python
import joblib

joblib.dump(model,"model2.pkl")
p_model=joblib.load("model2.pkl")
p_model.summary()

```
